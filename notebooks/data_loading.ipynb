```python
import pandas as pd

youtube_path = "../data/youtube_raw.csv"
gdelt_path = "../data/gdelt_raw.csv"
```

```python
yt = pd.read_csv(youtube_path)
print("YouTube raw shape:", yt.shape)
```

```python
gd = pd.read_csv(gdelt_path)
print("GDELT raw shape:", gd.shape)
```

```python
print("YouTube columns:", yt.columns.tolist())
yt.head()
```

```python
print("GDELT columns:", gd.columns.tolist())
gd.head()
```

```python
gd["published_at_dt"] = pd.to_datetime(
    gd["published_at"],
    format="%Y%m%dT%H%M%SZ",
    errors="coerce",
    utc=True
)
```

```python
print("YouTube info:")
yt.info()
print("\nMissing values (YouTube):")
print(yt.isna().sum())
```

```python
print("GDELT info:")
gd.info()
print("\nMissing values (GDELT):")
print(gd.isna().sum())
```

```python
yt["comment_published_at_dt"] = pd.to_datetime(
    yt["comment_published_at"], errors="coerce"
)

print("YouTube date range (comments):")
print("  min:", yt["comment_published_at_dt"].min())
print("  max:", yt["comment_published_at_dt"].max())
```

```python
print("GDELT date range (articles):")
print("  min:", gd["published_at_dt"].min())
print("  max:", gd["published_at_dt"].max())
```

```python
summary = pd.DataFrame([
    {
        "dataset": "youtube",
        "n_rows": len(yt),
        "n_columns": yt.shape[1],
        "n_missing_text": yt["comment_text"].isna().sum(),
        "date_min": yt["comment_published_at_dt"].min(),
        "date_max": yt["comment_published_at_dt"].max(),
    },
    {
        "dataset": "gdelt",
        "n_rows": len(gd),
        "n_columns": gd.shape[1],
        "n_missing_text": gd["title"].isna().sum(),
        "date_min": gd["published_at_dt"].min(),
        "date_max": gd["published_at_dt"].max(),
    },
])

summary
```

```python
summary_path = "../results/tables/data_overview.csv"
summary.to_csv(summary_path, index=False)
summary_path
```
